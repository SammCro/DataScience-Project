```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
list(air_poll.columns)
```

```python
new_df = air_poll.drop_duplicates(subset=["Country"])
```

```python
new_df
```

```python
categorical_columns = [col for col in new_df.columns if new_df[col].dtype in ["object"]]
```

```python
categorical_columns
```

```python
numerical_cols = [col for col in new_df.columns if col not in categorical_columns]
```

```python
numerical_cols
```

```python
categorical_columns = [
 'AQI Category',
 'CO AQI Category',
 'Ozone AQI Category',
 'NO2 AQI Category',
 'PM2.5 AQI Category']
```

```python
new_df=pd.get_dummies(new_df,columns=categorical_columns,drop_first=True)
```

```python
new_df.drop("City",axis=1,inplace=True)
```

```python
new_df
```

```python
df=new_df.drop("Country",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df=sc.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=5,init="k-means++",n_init=10,max_iter=200)
kmeans.fit(df)
pred = kmeans.predict(df)
```

```python
new_df["cluster"]=pred
```

```python
new_df.groupby("cluster").mean()
```
